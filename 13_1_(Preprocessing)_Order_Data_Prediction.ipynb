**B.DATA PREPROCESSING**

***1.Chỉnh sửa 2 cột 'ORDER_CREATION_DATE' và 'REQUESTED_DELIVERY_DATE':***




```python
#Chuyển 2 cột này sang dạng Datetime
data['ORDER_CREATION_DATE'] = pd.to_datetime(data['ORDER_CREATION_DATE'] , format='%Y%m%d')
data['REQUESTED_DELIVERY_DATE'] = pd.to_datetime(data['REQUESTED_DELIVERY_DATE'], format='%Y%m%d')
```

```python
#Kiểm tra xem có bao nhiêu dòng mà trong đó: Ngày tạo đơn hàng xảy ra sau ngày yêu cầu gửi hàng đến. Những dòng này là vô lý.
data[data['ORDER_CREATION_DATE'] > data['REQUESTED_DELIVERY_DATE']].shape[0]
```

```python
#Xóa bớt đi
data = data[data['ORDER_CREATION_DATE'] <= data['REQUESTED_DELIVERY_DATE']]
```

```python
#Thêm vào cột "Processing days" tính số ngày đơn hàng được giao
data.loc[:, 'Processing days'] = (data['REQUESTED_DELIVERY_DATE'] - data['ORDER_CREATION_DATE']).dt.days
```

***2.Sửa lại một số lỗi chính tả trong cột 'ORDER AMOUNT':***

```python
# Lọc và in ra các dòng chứa dấu '-', đây là những cột bị lỗi, cần sửa lại
negative_amount_rows = data[data['ORDER_AMOUNT'].str.contains('-')]
print(negative_amount_rows.shape)
```

```python
#Sửa lại các dòng này bằng cách xóa dấu '-'
data.loc[:, 'ORDER_AMOUNT'] = data['ORDER_AMOUNT'].str.replace('-', '')
```

```python
# Lọc và in ra các dòng chứa dấu ',', đây là những cột bị lỗi, cần sửa lại thành dấu '.'
error_amount_rows =data[data['ORDER_AMOUNT'].str.contains(',')]
print(error_amount_rows.shape)
```

```python
#Sửa lại các dòng này bằng cách xóa dấu '-'
data.loc[:, 'ORDER_AMOUNT'] = data['ORDER_AMOUNT'].str.replace(',', '.')
```

***3.Sửa lại một số lỗi chính tả trong cột 'ORDER_CURRENCY':***

```python
## Lọc và in ra các dòng ORDER_CURRENCY có chứa 'HU1', cần sửa lại thành 'HUI'
error_currency_rows =data[data['ORDER_CURRENCY'].str.contains('HU1')]
print(error_currency_rows.shape)
```

```python
#Sửa lại các dòng này
data.loc[:, 'ORDER_CURRENCY'] = data['ORDER_CURRENCY'].str.replace('HU1', 'HUI')
```

```python
data['ORDER_CURRENCY'].value_counts()
```

***Quy đổi giá tiền sang USD cho cột 'ORDER CURRENCY':***

```python
def convert_currency(amount, currency):
    if currency == 'USD':
        return amount
    else:
        url = 'https://www.google.com/search?q={} +{} +to +usd '.format(amount, currency)
        response = requests.get(url)
        soup = BeautifulSoup(response.text, 'html.parser')
        result = soup.find('div', attrs={'class':'BNeawe iBp4i AP7Wnd'}).text
        result = result.split(' ')[0]
        return result
```

```python
symbols = [x for x in data['ORDER_CURRENCY'].unique()]

dic = {}
for symbol in symbols:
    try:
        dic[symbol] = convert_currency('1', symbol)
    except:
        print(symbol)
```

```python
data = data[~data['ORDER_CURRENCY'].isin(['HUI', 'MYR'])]
```

```python
data.loc[:, 'amount_in_usd'] = data.apply(lambda x: float(x['ORDER_AMOUNT']) * float(dic[x['ORDER_CURRENCY']]), axis=1)
```

***4.Chỉnh sửa cột RELEASED_CREDIT_VALUE***

```python
# Thay đổi dấu phẩy thành dấu chấm trong cột 'RELEASED_CREDIT_VALUE'
data.loc[:, 'RELEASED_CREDIT_VALUE'] = data['RELEASED_CREDIT_VALUE'].str.replace(',', '.')
```

***5.Bỏ đi một số cột không cần thiết nữa***

```python
data=data.drop(columns=['CUSTOMER_ORDER_ID', 'ORDER_AMOUNT', 'ORDER_CURRENCY',
                        'ORDER_CREATION_TIME', 'ORDER_CREATION_DATE', 'REQUESTED_DELIVERY_DATE'])
```
